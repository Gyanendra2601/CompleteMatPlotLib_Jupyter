# Getting Started with MatplotLib

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,4,7],[45,76,21,34])
plt.show()
```

```python
x= [1,32,40,76,89,90]
y = [54,76,89,90,65,98]

x2= [1,32,45,76,89,95]
y2 = [54,76,89,70,65,98]

plt.plot(x,y, label = 'First Line')
plt.plot(x2,y2, label = 'Second line')
plt.xlabel('X Xaxes')
plt.ylabel('Y axes')
plt.legend()
plt.title("This is the\nTitle")
plt.show()
```

```python
x= [2,4,6,8,10,12]
y = [54,76,89,90,65,98]

x2= [3,5,7,9,11,13]
y2 = [54,76,89,70,65,98]

plt.bar(x,y, label = 'First Line',color = 'r')
plt.bar(x2,y2, label = 'Second line', color = 'k')
plt.xlabel('X Xaxes')
plt.ylabel('Y axes')
plt.legend()
plt.title("This is the\nTitle")
plt.show()
```

```python
Ages= [24,34,65,78,89,90,12,67,34,56,76,45,67,45,43,78,87]

Ids = [x for x in range(len(Ages))]

bins = [10,20,30,40,50,60,70,80,90]
plt.hist(Ages,bins,histtype = 'bar',width = 8, color = 'g')
plt.xlabel('X Xaxes')
plt.ylabel('Y axes')
#plt.legend()
plt.title("This is the\nTitle")
plt.show()
```
